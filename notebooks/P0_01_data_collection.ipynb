### 3. Load Dataset
Load the "BESSTIE" dataset from the Hugging Face Hub. This dataset will be downloaded and stored in the `dataset` variable.

```python
from datasets import load_dataset

dataset = load_dataset("unswnlporg/BESSTIE")
dataset
```

### 4. Save Data
Define the directory path for saving the raw data. Create the directory if it doesn't exist, and then save the 'train' and 'validation' splits of the dataset as CSV files to the specified location.

```python
import os

save_path = "/content/drive/MyDrive/DNLP/data/raw"
os.makedirs(save_path, exist_ok=True)

dataset["train"].to_csv(f"{save_path}/besstie_train.csv", index=False)
dataset["validation"].to_csv(f"{save_path}/besstie_validation.csv", index=False)

print("Saved to:", save_path)
```
